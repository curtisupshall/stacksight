````python
import scipy

def get_score(a, b):
    return 1 - scipy.spatial.distance.cosine(a, b)
    # return dot(a, b)
````

````python
def generate_embeddings(sentences):
    return [(sentence, get_embedding(sentence)) for sentence in sentences]
````

````python
def score_embeddings(embedding_pairs, keyword):
    keyword_embedding = get_embedding(keyword)
    scored_embeddings = [(sentence, get_score(embedding, keyword_embedding)) for (sentence, embedding) in embedding_pairs]
    return sorted(scored_embeddings, key=lambda x: x[1], reverse=True)
````
